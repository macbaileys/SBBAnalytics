## Average traffic (DTV, DVW, DNWV) per canton

```python
# Calculate the average traffic (DTV, DVW, DNWV) per canton
select_query_avg_traffic = """
SELECT Kanton, 
       AVG(DTV) AS avg_DTV, 
       AVG(DWV) AS avg_DWV, 
       AVG(DNWV) AS avg_DNWV 
FROM data_prepared 
GROUP BY Kanton;
"""
cursor.execute(select_query_avg_traffic)
avg_traffic = cursor.fetchall()

if avg_traffic:
    for row in avg_traffic:
        print(row)
else:
    print("No results for average traffic per canton")
```

## Stations with the highest total daily traffic in 2023

```python
# 3. Calculate the station with the highest total daily traffic in 2023
select_query_max_daily_traffic = """
SELECT Bahnhof, DTV AS Total_Daily_Traffic 
FROM data_prepared 
WHERE Jahr = %s 
ORDER BY Total_Daily_Traffic DESC 
LIMIT 1;
"""
cursor.execute(select_query_max_daily_traffic, (2023,))
max_daily_traffic = cursor.fetchone()

if max_daily_traffic:
    print("Station with highest daily traffic in 2023:", max_daily_traffic)
else:
    print("No results for the station with the highest daily traffic in 2023")
```

## Stations operated by each EVU

```python
# 4. Count the number of stations operated by each EVU
select_query_evu_count = """
SELECT EVU, COUNT(*) AS station_count 
FROM data_prepared 
GROUP BY EVU 
ORDER BY station_count DESC;
"""
cursor.execute(select_query_evu_count)
evu_counts = cursor.fetchall()

if evu_counts:
    for row in evu_counts:
        print(row)
else:
    print("No results for EVU station counts")
```

## Query all stations within a specific geographical bounding box


```python
# 5. Retrieve all stations within a specific geographical bounding box
select_query_geo_box = """
SELECT * 
FROM data_prepared 
WHERE lat BETWEEN %s AND %s AND lon BETWEEN %s AND %s;
"""
cursor.execute(select_query_geo_box, (46, 48, 6, 10))
geographical_stations = cursor.fetchall()

if geographical_stations:
    for row in geographical_stations:
        print(row)
else:
    print("No results for stations within the geographical bounding box")
```
